```python
from fastai.collab import *
from fastai.tabular.all import *
path = untar_data(URLs.ML_100k)
```

```python
# Initialize the data
ratings = pd.read_csv(path/'u.data', delimiter='\t', header=None,
                      names=['user','movie','rating','timestamp'])
ratings.head()
movies = pd.read_csv(path/'u.item',  delimiter='|', encoding='latin-1',
                     usecols=(0, 1), names=('movie', 'title'), header=None)
movies.head()

ratings = ratings.merge(movies)
ratings.head()

```

```python
# Create DataLoader
dls = CollabDataLoaders.from_df(ratings, item_name='title', bs=64)
dls.show_batch()
```

```python
learn = collab_learner(dls, n_factors=50, y_range=(0, 5.5))
```

```python
learn.fit_one_cycle(5, 5e-3, wd=0.1)
learn.save('trained_model')
```

```python
import re
def convert_book_name(book_name: str):
    # Get rid of parts/series etc.
    prev_name = book_name
    book_name = "".join(
        filter(lambda chr: chr.isalpha() or chr in ' ', book_name))
    
    book_name = re.sub(' +', ' ', book_name)
    return book_name.rstrip()
convert_book_name('Hello 1')
```

```python
def replace_from_set(string: str, regex_set: [str]):
    for regex_search_term in regex_set:
        string = re.sub(regex_search_term, "", string)
    return string


replace_from_set('Иные Город полный загадок Глава 3 СИ', ['Глава \d.*'])
```

```python
class Singleton(type):
    _instances = {}

    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super(
                Singleton, cls).__call__(*args, **kwargs)
        return cls._instances[cls]



class SomeClass(metaclass=Singleton):
    def __init__(self,x):
        super().__init__()
        self.x = x
        print("Initiing SomeClass")
    def method(self):
        self.x+=5
        print(self.x)

x = SomeClass(4)
x.method()

y = SomeClass()
y.method()
```
